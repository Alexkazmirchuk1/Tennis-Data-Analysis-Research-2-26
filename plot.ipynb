```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

# Calculate the probability of holding serve
def prob_hold(p):
    """Probability server holds."""
    q = 1 - p
    return (
        p**4 +
        4 * (p**4) * q + 
        10 * (p**4) * (q**2) + 
        (20 * (p**3) * (q**3)) * ((p**2) / (1 - 2 * p * q))
    )

# Calculate the probability of winning a game given the probability of winning a point
def prob_win_game(p):
    """Probability of winning a game given point winning probability."""
    q = 1 - p
    return (
        p**4 +                          # Player wins the game on first four points
        4 * (p**4) * q +                # Player wins the game with one point loss
        10 * (p**4) * (q**2) +          # Player wins the game with two points lost
        20 * (p**3) * (q**3) * (p**2) / (1 - 2 * p * q)  # Player wins after losing three points
    )

# Calculate the probability of winning a set given the probability of winning a game
def prob_win_set(ps):
    """Probability of winning a set."""
    qs = 1 - ps
    return (
        ps**6 +
        6 * (ps**6) * qs + 
        21 * (ps**6) * (qs**2) + 
        56 * (ps**6) * (qs**3) +
        126 * (ps**6) * (qs**4) +
        42 * (ps**7) * (qs**5) +
        924 * (ps**7) * (qs**6)
    )

# Calculate the probability of winning a match given the probability of winning a set
def prob_win_match(pm):
    """Probability of winning a match."""
    qm = 1 - pm
    return (
        pm**3 +
        3 * (pm**3) * qm + 
        6 * (pm**3) * (qm**2)
    )

# Generate an array of probabilities of winning a point
point_win_probs = np.linspace(0, 1, 100)
game_probs = []
match_probs = []

# Calculate the probabilities for games and matches for each point winning probability
for p in point_win_probs:
    game_probability = prob_win_game(p)
    set_probability = prob_win_set(game_probability)
    match_probability = prob_win_match(set_probability)

    game_probs.append(game_probability)
    match_probs.append(match_probability)

# Orange is left chart, blue is right chart
orange_line = '#FF9500'
blue_line = '#5BADFF'

# Generate figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))

# Probability of Winning a Game
ax1.plot(point_win_probs, game_probs, label='Probability of Winning Game', color=orange_line, lw=2)
ax1.set_title('Winning a Game', fontsize=16)
ax1.set_xlabel('Probability of Winning Each Point', fontsize=14)
ax1.set_ylabel('Probability of Winning Game', fontsize=14)
ax1.grid(linestyle='--', alpha=0.7)
ax1.legend(fontsize=12)
ax1.set_ylim(0, 1)
ax1.set_xlim(0, 1)

# Probability of Winning a Match
ax2.plot(point_win_probs, match_probs, label='Probability of Winning Match', color=blue_line, lw=2)
ax2.set_title('Winning a Match', fontsize=16)
ax2.set_xlabel('Probability of Winning Each Point', fontsize=14)
ax2.set_ylabel('Probability of Winning Match', fontsize=14)
ax2.grid(linestyle='--', alpha=0.7)
ax2.legend(fontsize=12)
ax2.set_ylim(0, 1)
ax2.set_xlim(0, 1)

plt.tight_layout(pad=3.0)
plt.show()
```
